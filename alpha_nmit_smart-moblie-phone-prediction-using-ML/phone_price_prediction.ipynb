```python
# Import libraries
import numpy as np #This line imports the NumPy library and makes it available under the alias "np"
import pandas as pd #This line imports the Pandas library and makes it available under the alias "pd"
import seaborn as sns #This line imports the Seaborn library, a data visualization library in Python
from matplotlib import pyplot as plt #This line imports the pyplot module from the Matplotlib library and makes it available under the alias "plt"
# Import Dataset
df = pd.read_csv("/content/data.csv") #The first line reads a CSV file named "data.csv" located at the path "/content/data.csv" and assigns the data to a variable called "df"
df.head() #The second line is a method call on the DataFrame object "df"
# Drop the Unwannted column the first one
df.drop(['Unnamed: 0'],axis=1,inplace=True) #removes the column 'Unnamed: 0' from the DataFrame 'df' permanently
# After removing the dataset look like
df.head() # display the first few rows
# Checking null values into the dataset
df.isnull().sum() #count the number of missing values in each column
# Seeing the null values using heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis') #create a heatmap visualization of the missing values
# Summary of the dataset
df.info() # provides a concise summary
df.dtypes # displays the data type of each column
# Describe
df.describe()
# Dropping the Name columns
df.drop(['Brand me'],axis=1,inplace=True)
# We should handling these null or missing values
df.isnull().sum()
# Fill up the mean values of all the missing value columns into the dataset
df['Ratings'] = df['Ratings'].fillna(df['Ratings'].mean())
df['RAM'] = df['RAM'].fillna(df['RAM'].mean())
df['ROM'] = df['ROM'].fillna(df['ROM'].mean())
df['Mobile_Size'] = df['Mobile_Size'].fillna(df['Mobile_Size'].mean())
df['Selfi_Cam'] = df['Selfi_Cam'].fillna(df['Selfi_Cam'].mean())
# After handling the all of the missing and null values from the dataset
df.isnull().sum() # used to calculate the sum of missing values in each column

sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis') #create a heatmap visualization to highlight the missing values using the Seaborn library.
# Changing the Datatype
df['RAM'] = df['RAM'].astype('int64') #convert the data type of the 'RAM' column to 'int64'.
df['ROM'] = df['ROM'].astype('int64') # converts ROM
df['Selfi_Cam'] = df['Selfi_Cam'].astype('int64') # converts data type of 'selfie cam' column into 64
# Final dataset for build a model
df.head()
# Finding out the correlation between the features
corr = df.corr() # calculates the correlation matrix
corr.shape #  access the shape attribute of the correlation matrix
# Plotting the heatmap of correlation between features
plt.figure(figsize=(12,12))
sns.heatmap(corr, cbar=False, square= True, fmt='.2%', annot=True, cmap='Greens')
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='Ratings',data=df)
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='RAM',data=df)
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='ROM',data=df)
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='Primary_Cam',data=df)
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.countplot(x='Selfi_Cam',data=df)
sns.distplot(df['RAM'].dropna(),kde=False,color='darkred',bins=10)
sns.distplot(df['Battery_Power'].dropna(),kde=False,color='green',bins=10)
sns.distplot(df['Price'].dropna(),kde=False,color='darkblue',bins=15)
sns.distplot(df['Battery_Power'].dropna(),kde=False,color='darkblue',bins=15)
plt.figure(figsize=(10,10))
sns.pairplot(data=df)
# Lets try to understand which are important feature for this dataset
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X = df.iloc[:,1:7]  # Independent columns
y = df.iloc[:,[-1]] # Yarget column i.e price range
# Apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=4)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
# Concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
featureScores
print(featureScores.nlargest(4,'Score'))  #print 5 best features
# Fiting Feature Seclection using Ensemble Methods
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
# Plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
df.head()
# Value Assigning
X = df.iloc[:,[6,2,4,5,1,3]]
y = df.iloc[:,[-1]]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_train)
# Training Score
print("Training Accuracy:",reg.score(X_train,y_train)*100)
# Testing Score
print("Testing Accuracy:",reg.score(X_test,y_test)*100)
# Visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
plt.show()
# Sample Prediction
reg.predict([[4.0,128.0,6.00,48,13.0,4000]])
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=15)
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
y_pred = reg.predict(X_train)
# Training Score
print("Training Accuracy:",reg.score(X_train,y_train)*100)
# Testing Score
print("Testing Accuracy:",reg.score(X_test,y_test)*100)
# Sample Prediction
reg.predict([[4.0,128.0,6.00,48,13.0,4000]])
```
